```python
data1 = pd.read_csv('google-playstore-apps/Google-Playstore-Transformed.csv.gz',compression='gzip')
data1
```

```python
data1.drop(['Min_Downloads'],1,inplace=True)
```

```python
data1 = pd.get_dummies(data1)
data1
```

```python
y = data1[['Average_Downloads']]
```

```python
X = data1.drop(['Average_Downloads'],axis=1)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)
```

```python
print(X_train.shape,X_test.shape,y_train.shape,y_test.shape)
```

```python
from catboost import CatBoostRegressor
from sklearn.metrics import mean_absolute_error
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import KFold

catboost_model = CatBoostRegressor()
catboost_model.fit(X_train, y_train)
predicted_catboost = catboost_model.predict(X_test)
print("Mean Absolute Error using cat boost: %.3f" % mean_absolute_error(y_test, predicted_catboost))
```

```python
from catboost import CatBoostRegressor
from sklearn.metrics import mean_absolute_error
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import KFold
catboost_model = CatBoostRegressor(depth=9, iterations = 800, learning_rate = 0.31)

catboost_model.fit(X_train, y_train)
predicted_catboost = catboost_model.predict(X_test)
print("Mean Absolute Error using cat boost grid: %.3f" % mean_absolute_error(y_test, predicted_catboost))
```
