```python
import pandas as pd
import numpy as np 
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
data1 = pd.read_csv('Playstore-data1.csv.gz',compression='gzip')
data1
```

```python
data1.drop(['Unnamed: 0','Average_Downloads'],1,inplace=True)
data1.head()
```

```python
data1 = pd.get_dummies(data1)
data1
```

```python
y = data1[['Min_Downloads']]
y
```

```python
X = data1.drop(['Min_Downloads'],axis=1)
X
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)
```

```python
print(X_train.shape,X_test.shape,y_train.shape,y_test.shape)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier()
param_grid = {"n_estimators": [50, 100, 200], "max_depth": [20, 30, 40, 50], "max_features": ['sqrt']}


grid_search = GridSearchCV(rf, param_grid=param_grid)
grid_search.fit(X_train, y_train)

print("Best score is {}".format(grid_search.best_params_))
print("Best score is {}".format(grid_search.best_score_))
# Best parameters are {'max_depth': 20, 'max_features': 'sqrt', 'n_estimators': 200}
```
