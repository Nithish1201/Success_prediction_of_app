```python
data1 = pd.read_csv('google-playstore-apps/Google-Playstore-Transformed.csv.gz',compression='gzip')
data1
```

```python
data1.drop(['Min_Downloads'],1,inplace=True)
data1.head()
```

```python
data1 = pd.get_dummies(data1)
data1
```

```python
y = data1[['Average_Downloads']]
y
```

```python
X = data1.drop(['Average_Downloads'],axis=1)
X
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)
```

```python
print(X_train.shape,X_test.shape,y_train.shape,y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error
linear_reg_model = LinearRegression()
linear_reg_model.fit(X_train, y_train)
yhat = linear_reg_model.predict(X_test)
mae = mean_absolute_error(y_test, yhat)
print('MAE: %.3f' % mae)
```

```python
from sklearn.tree import DecisionTreeRegressor
decision_model = DecisionTreeRegressor()  
decision_model.fit(X_train, y_train) 
predicted_decision_trees = decision_model.predict(X_test)
print ("Mean Absolute Error using Decision Tress : %.3f" % mean_absolute_error(y_test, predicted_decision_trees))
```

```python
from sklearn.ensemble import RandomForestRegressor
forest_model = RandomForestRegressor(n_estimators=100, max_depth=10)
forest_model.fit(X_train, y_train.values.ravel())
predicted_random_forest = forest_model.predict(X_test)
print("Mean Absolute Error using Random Forest: %.3f" % mean_absolute_error(y_test, predicted_random_forest))
```

```python
from xgboost import XGBRegressor
xg_model = XGBRegressor(n_estimators=100)
xg_model.fit(X_train, y_train)
predicted_XGBoost = xg_model.predict(X_test)
print("Mean Absolute Error using XGBoost: %.3f" % mean_absolute_error(y_test, predicted_XGBoost))
```

```python
from sklearn.linear_model import SGDRegressor
from sklearn.metrics import mean_absolute_error
SGD_model =SGDRegressor(max_iter = 1000,penalty = "elasticnet",loss = 'huber',tol = 1e-3, average = True)
SGD_model.fit(X_train, y_train.values.ravel())
predicted_SGD = SGD_model.predict(X_test)
print("Mean Absolute Error using SGD: %.3f" % mean_absolute_error(y_test, predicted_SGD))
```

```python
from sklearn.linear_model import ElasticNet
elasticNet_model = ElasticNet(alpha=1.0, l1_ratio=0.5)
elasticNet_model.fit(X_train, y_train)
predicted_elasticNet = elasticNet_model.predict(X_test)
print("Mean Absolute Error using Elastic net: %.3f" % mean_absolute_error(y_test, predicted_elasticNet))
```

```python
from sklearn.ensemble import GradientBoostingRegressor
GDboost_model = GradientBoostingRegressor()
GDboost_model.fit(X_train, y_train.values.ravel())
predicted_GDboost = GDboost_model.predict(X_test)
print("Mean Absolute Error using GD boost: %.3f" % mean_absolute_error(y_test, predicted_GDboost))
```

```python
from sklearn.linear_model import BayesianRidge
from sklearn.metrics import mean_absolute_error
bayes_model = BayesianRidge()
bayes_model.fit(X_train, y_train.values.ravel())
predicted_bayes = bayes_model.predict(X_test)
print("Mean Absolute Error using Bayes rigde: %.3f" % mean_absolute_error(y_test, predicted_bayes))
```

```python
from lightgbm import LGBMRegressor
from sklearn.metrics import mean_absolute_error
LGBM_model = LGBMRegressor()
LGBM_model.fit(X_train, y_train)
predicted_LGBM = LGBM_model.predict(X_test)
print("Mean Absolute Error using LGBM: %.3f" % mean_absolute_error(y_test, predicted_LGBM))
```

```python
from sklearn.cross_decomposition import PLSRegression
from sklearn.metrics import mean_absolute_error
PLS_model = PLSRegression()
PLS_model.fit(X_train, y_train)
predicted_PLS = PLS_model.predict(X_test)
print("Mean Absolute Error using PLS: %.3f" % mean_absolute_error(y_test, predicted_PLS))
```

```python
from catboost import CatBoostRegressor
from sklearn.metrics import mean_absolute_error
catboost_model = CatBoostRegressor()
catboost_model.fit(X_train, y_train)
predicted_catboost = catboost_model.predict(X_test)
print("Mean Absolute Error using cat boost: %.3f" % mean_absolute_error(y_test, predicted_catboost))
```

```python
from sklearn.linear_model import Ridge
from sklearn.metrics import mean_absolute_error
ridge_model = Ridge(alpha = 0.1)
ridge_model.fit(X_train, y_train)
predicted_ridge = ridge_model.predict(X_test)
print("Mean Absolute Error using Ridge: %.3f" % mean_absolute_error(y_test, predicted_ridge))
```
