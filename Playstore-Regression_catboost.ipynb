```python
import pandas as pd
import numpy as np 
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
data1 = pd.read_csv('Playstore-data1.csv.gz',compression='gzip')
data1
```

```python
data1.drop(['Unnamed: 0','Min_Downloads'],1,inplace=True)
data1.head()
```

```python
data1 = pd.get_dummies(data1)
data1
```

```python
y = data1[['Average_Downloads']]
y
```

```python
X = data1.drop(['Average_Downloads'],axis=1)
X
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)
```

```python
print(X_train.shape,X_test.shape,y_train.shape,y_test.shape)
```

```python
from catboost import CatBoostRegressor
from sklearn.metrics import mean_absolute_error
catboost_model = CatBoostRegressor()
catboost_model.fit(X_train, y_train)
predicted_catboost = catboost_model.predict(X_test)
print("Mean Absolute Error using cat boost: %.3f" % mean_absolute_error(y_test, predicted_catboost))
```

```python
from catboost import CatBoostRegressor
from sklearn.metrics import mean_absolute_error
#catboost_model_grid = CatBoostRegressor(depth=9, iterations = 500, learning_rate = 0.2)
catboost_model_grid = CatBoostRegressor(depth=9, iterations = 800, learning_rate = 0.31)
catboost_model_grid.fit(X_train, y_train)
predicted_catboost_grid = catboost_model_grid.predict(X_test)
print("Mean Absolute Error using cat boost grid: %.3f" % mean_absolute_error(y_test, predicted_catboost_grid))
```
